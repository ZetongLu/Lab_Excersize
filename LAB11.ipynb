```python
import requests
import json
import random
from tkinter import *

def get_question():
    inform = requests.get(f"https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy")
    json_data = inform.json()
    question_number = random.randint(0, 9)

    question = json_data[question_number]["question"]
    correct = json_data[question_number]["correctAnswer"]
    incorrect = json_data[question_number]["incorrectAnswers"]
    return question, correct, incorrect

def correct_click():
    global score
    score += 1
    label4.config(text="")
    next_question()

def incorrect_click():
    label4.config(text="You are wrong....")
    next_question()

def next_question():
    question, correct, incorrect = get_question()
    my_label1.config(text=question)
    button1.config(text=correct, command=correct_click)
    button2.config(text=incorrect, command=incorrect_click)
    root.after(10000, show_score)  # 10 seconds to answer each question

def show_score():
    label3.config(text=f"Your score: {score}")

score = 0
root = Tk()
root.geometry("500x500")

question, correct, incorrect = get_question()

my_label1 = Label(root, text=question)
my_label1.grid(row=0, column=0)

button1 = Radiobutton(root, text=correct, command=correct_click, padx=20, pady=20)
button2 = Radiobutton(root, text=incorrect, command=incorrect_click, padx=20, pady=20)

button1.grid(row=3, column=0)
button2.grid(row=4, column=0)

label3 = Label(root, text="")
label3.grid(row=5, column=0)

label4 = Label(root, text="")
label4.grid(row=6, column=0)

root.after(10000, show_score)
root.mainloop()
```
