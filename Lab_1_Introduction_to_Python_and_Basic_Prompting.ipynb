# PART 3: Text Analysis - Understanding Code

Now we'll analyze a text about surveillance capitalism.
This is the kind of thing computers do all the time - count words, find patterns.

## We're going to:
1. Take a text
2. Count how many words it has
3. Search for specific terms related to our course
4. Display the results

## Exercise 3.1: Analyzing Text (READ THIS CODE)

Below is a text excerpt about surveillance capitalism.

**READ THE CODE SLOWLY** and follow the comments (lines starting with `#`)

````python
# This is our text to analyze - an excerpt about surveillance capitalism
text = """
Surveillance capitalism is an economic system centered around the
commodification of personal data with the core purpose of profit-making.
Data is extracted from users through various digital platforms and devices.
This data is then analyzed using algorithms and artificial intelligence
to predict and influence human behavior. The surveillance capitalist model
relies on the accumulation of vast amounts of behavioral data, which is
converted into predictions that are sold to businesses for targeted advertising.
"""

# STEP 1: Split the text into individual words
# .split() breaks text at spaces, creating a list of words
words = text.split()

# STEP 2: Count how many words we have
# len() means "length" - how many items in the list
word_count = len(words)

# STEP 3: Display the result
print("=" * 50)
print("TEXT ANALYSIS RESULTS")
print("=" * 50)
print(f"Total words in text: {word_count}")
print()
````

## Exercise 3.2: Searching for Terms

Now let's search for specific terms related to surveillance capitalism.

### The code below:
1. Defines a list of terms we're looking for
2. Looks through each word in the text
3. Checks if that word matches our search terms
4. Counts how many times we found each term

**Run the cell below to see the results:**

````python
# Terms we want to find (related to Unit 1 themes)
search_terms = ["surveillance", "data", "algorithms", "behavior", "profit"]

# This will store the terms we actually find
found_terms = []

# Look through each word in the text
for word in words:
    # Clean up the word: lowercase and remove punctuation
    clean_word = word.lower().strip('.,!?')

    # Check if this word is in our search terms
    if clean_word in search_terms:
        found_terms.append(clean_word)

# Display results
print(f"Search terms found: {len(found_terms)} times")
print(f"Terms found: {found_terms}")
print()

# Count how many times each term appears
print("Frequency of each term:")
for term in search_terms:
    count = found_terms.count(term)
    if count > 0:
        print(f"  '{term}': {count} time(s)")
````
