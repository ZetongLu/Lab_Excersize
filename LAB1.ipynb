```python
#hangman program in function, final verstion
import time
word = "hangman"

def listletter(word):
    list_word = {}
    for i in word:
        if i not in list_word.keys():
            list_word[i] = 1
        else:
            list_word[i] += 1
    return list_word

def create_emtylist(word):
    emtylist = []
    for l in word:
        emtylist.append("_")
    return emtylist

def guesslistchange(guesslist,letter):
    guesslist.append(letter)
    return guesslist

def visualize(word,emtylist,letter):
    count = 0
    for m in word:
        if letter == m:
            emtylist[count] = m
        count += 1
    time.sleep(1)
    return emtylist
    

def checkletter(word,letter,list_word,guesslist,life,emtylist):

    if letter in list_word.keys() and letter not in guesslist:
        list_word.pop(letter)
        guesslistchange(guesslist,letter)
        print("you got one correct")
        visualize(word,emtylist,letter)

    elif letter in list_word.keys() and letter in guesslist:
        print("Try another one, You've already tried this letter")
        
    elif letter not in list_word.keys() and letter in guesslist:
        print("Dont fall on the same mistake twice try another letter")

    else:
        print("Sorry... wrong!!!")
        guesslistchange(guesslist,letter)
        life -= 1

    print(life)
    return life
        
def checklife(life):
    if life <= 0:
        print("you lose")
        quit()
   
def main(word):
    life = 8
    guesslist = []
    list_word = listletter(word)
    emtylist = create_emtylist(word)
    print("this is a game of hangman fill up the colume by guessing letters")
    print(emtylist)
    time.sleep(1)
    while list_word != {}:
        print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
        print("YOU HAVE",life,"LIFES LEFT")
        letter = input("guess a letter")
        life = checkletter(word,letter,list_word,guesslist,life,emtylist)
        checklife(life)
        print(visualize(word,emtylist,letter))
    print("=-=-=-=-=-=-=-=-=-=-=-=")
    print("YOU GOT THE WORD!!!!")
    print("-=-=-=-=-=-=-=-==-=-=-=-=-=-=")

main(word)
```
