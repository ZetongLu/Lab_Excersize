```python
import requests

api_key = "7928c0363717466f82988bb03c4928be"

def get_top_headlines(country="us"):
    url = f"https://newsapi.org/v2/top-headlines?country={country}&apiKey={api_key}"
    response = requests.get(url)
    return response.json()

def search_news(keyword):
    url = f"https://newsapi.org/v2/everything?q={keyword}&apiKey={api_key}"
    response = requests.get(url)
    return response.json()

def headlines_by_source(source):
    url = f"https://newsapi.org/v2/top-headlines?sources={source}&apiKey={api_key}"
    response = requests.get(url)
    return response.json()

def list_sources():
    url = f"https://newsapi.org/v2/sources?apiKey={api_key}"
    response = requests.get(url)
    return response.json()

def get_source_id(source_name):
    sources = list_sources()["sources"]
    for source in sources:
        if source["name"].lower() == source_name.lower():
            return source["id"]
    return None

if __name__ == "__main__":
    print("Top US headlines:")
    top_headlines = get_top_headlines()
    for i, article in enumerate(top_headlines["articles"]):
        print(f"{i + 1}. {article['title']}")

    print("\nNews articles with keyword 'technology':")
    tech_news = search_news("technology")
    for i, article in enumerate(tech_news["articles"]):
        print(f"{i + 1}. {article['title']}")

    source_id = get_source_id("TechCrunch")
    if source_id:
        print(f"\nTop headlines from TechCrunch:")
        tc_headlines = headlines_by_source(source_id)
        for i, article in enumerate(tc_headlines["articles"]):
            print(f"{i + 1}. {article['title']}")
```
