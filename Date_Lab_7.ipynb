```python
import requests
import matplotlib.pyplot as plt

year = "2023"
country_codes = ["US", "CA", "MX", "GB", "DE", "FR", "IT", "JP", "AU", "BR"]

holidays_count = {}

for country_code in country_codes:
    response = requests.get(f"https://date.nager.at/api/v3/PublicHolidays/{year}/{country_code}")
    
    if response.status_code == 200:
        holidays = response.json()
        holidays_count[country_code] = len(holidays)
    else:
        print(f"Error fetching data for {country_code}, status code: {response.status_code}")

total_holidays = sum(holidays_count.values())

print("Total number of holidays:", total_holidays)
print("Holidays per country:", holidays_count)


country_codes = list(holidays_count.keys())
holidays_numbers = list(holidays_count.values())

plt.bar(country_codes, holidays_numbers)
plt.xlabel("Countries")
plt.ylabel("Number of Public Holidays")
plt.title("Public Holidays in 10 Countries")
plt.show()
```
